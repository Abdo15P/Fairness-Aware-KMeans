```python
def balance_measure(cluster_combinations_perc, global_combinations_perc):

    #Calculate difference between cluster combination percentages and global combination percentages
    deviations = []
    k = cluster_combinations_perc.shape[0]
    for i in range(k):
        for j in range(cluster_combinations_perc.shape[1]):
            deviations.append(abs(cluster_combinations_perc[i, j] - global_combinations_perc[j]))

    print(" ")
    print("balance measure")
    return deviations
```

```python
def fairness_measure(cluster_combinations_counts, cluster_combinations_perc, global_combinations_perc, k):
    clusters_fairness = []
    data_size = np.sum(cluster_combinations_counts)
    #Calculate the total deviations of all clusters, each cluster deviation by its weight (cluster size/total size)
    for i in range(k):
        fairness_value = 0
        for j in range(global_combinations_perc.shape[0]):
            fairness_value = fairness_value + abs(cluster_combinations_perc[i, j] - global_combinations_perc[j])/100
        fairness_value = fairness_value * ((np.sum(cluster_combinations_counts[i])) / data_size)
        clusters_fairness.append(fairness_value)
    return sum(clusters_fairness)
```

```python
def cluster_quality_measure(data, centroids, cluster_count, cluster_label):
    bcss = 0 #between‑cluster sum of squares, measure of how far apart the centroids are
    wcss = 0 #within‑cluster sum of squares, total squared distance of each point to its own centroid.
    for i in range(len(centroids)):
        current_cluster_size = np.sum(cluster_count[i])
      #calculate differrence between centroids and multiply by cluster size
        for j in range(i+1, len(centroids)):
            other_cluster_size = np.sum(cluster_count[j])
            w = (current_cluster_size * other_cluster_size) / data.shape[0]
            d1 = w * np.sum((centroids[i] - centroids[j])**2)
            bcss = bcss + d1

        cluster_data = data[cluster_label == i]
        #calculate distance betwwen each pont and its cluster centroid
        for k in range(cluster_data.shape[0]):
            d2 = np.sum((cluster_data[k] - centroids[i])**2)
            wcss = wcss + d2
    return bcss, wcss
```
