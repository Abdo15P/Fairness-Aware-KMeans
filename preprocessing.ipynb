```python
import pandas as pd #for reading and manipulating data
from sklearn.compose import ColumnTransformer # preprocessing pipelines to different subsets of columns
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Public URL of the Adult dataset
#DATA_URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

def preprocessing():
    column_headers = [
        "age", "workclass", "fnlwgt", "education", "education-num",
        "marital-status", "occupation", "relationship", "race", "sex",
        "capital-gain", "capital-loss", "hours-per-week", "native-country",
        "income"
    ] #defining column headers for the dataset

    #Read file, set column headers, treat ? as missing values, remove space after commas
    df = pd.read_csv("adult.data", names=column_headers, na_values="?", skipinitialspace=True)


    # print(df.isna().any(axis=1).sum())  #no. of rows with missing values
    print("Before dropping rows with missing values:", len(df)) #no. of rows

    #Drop any row that contains at least one missing value
    df = df.dropna()
    print("After dropping rows with missing values:", len(df))

    sensitive_features = df[["race", "sex"]].to_numpy() #2D array of strings
    income = df["income"].to_numpy() #target variable, 1D array
    df = df.drop(columns=["fnlwgt", "race", "sex", "income"]) #drop sensitive features, irrelevant features and target


    numeric_features = [
        "age", "education-num", "capital-gain",
        "capital-loss", "hours-per-week"
    ] #list of continuous features

    categorical_features = [
        "workclass", "education", "marital-status",
        "occupation", "relationship", "native-country"
    ]#list of categorical features


    preprocessor = ColumnTransformer(
        transformers=[
            ("num", StandardScaler(), numeric_features),#prevent large variables from overpowering small ones.
            ("cat", OneHotEncoder(), categorical_features)
        ]
    )#apply different piplines to different columns

    df_transformed = preprocessor.fit_transform(df)
    data = df_transformed.toarray()

    #Debugging: uncomment to print demographic statistics

    #num_white = (sensitive_features[:, 0] == "White").sum()
    #num_male = (sensitive_features[:, 1] == "Male").sum()
    #print("num of white: ", num_white)
    #print("num of males: ", num_male)
    #print("white percentage: ", num_white / data.shape[0])
    #print("males percentage: ", num_male / data.shape[0])
    return data, sensitive_features, income
```
